```python
import numpy as np
import tensorly as tl
from tensorly import tucker_to_tensor
```

```python
x = tl.tensor(np.arange(64*64*20*20*2).reshape((64, 64, 20, 20, 2)))
```

```python
y = tl.tensor(np.arange(64*64*20*20*2).reshape((64, 64, 20, 20, 2)))
```

```python
from tensorly.decomposition import tucker
core, factors = tucker(x, rank=[20, 20, 20, 20, 2])
core.shape
```

```python
len(factors)
```

```python
for i in factors:
    print(i.shape)
```

```python
y = factors[0] @ x
y.shape
```

```python
z = factors[1] @ y
z
```

```python
tl.tucker_to_tensor((core, factors))
```
